## (1) $x=3$일 때 $x^2$과 $x^3$을 계산하고 출력하시오. 

```python
x=3
x2 = x**2
x3 = x**3
print(f"x^2={x2}, x^3={x3}")
```

## (2) y=log x 함수에 대한 그래프를 $x\in\{1,2,3,...,10\}$에 대해서 그리시오. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(1,11,1)
y = np.log(x)

plt.plot(x,y)
plt.show()
```
